### Range

```python
# Imprimindo números pares entre 50 e 101
for i in range(50, 101, 2):
    print(i)
```

```python
for i in range(3, 6):
    print (i)
```

```python
for i in range(0, -20, -2):
    print(i)
```

```python
lista = ['Morango', 'Banana', 'Abacaxi', 'Uva']
lista_tamanho = len(lista)
for i in range(0, lista_tamanho):
    print(lista[i])
```
