### Manipulando Arquivos CSV (comma-separated values )

```python
# Importando o módulo csv
import csv
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/PythonFundamentos-master/Cap04/Notebooks/arquivos/numeros.csv','w') as arquivo:
    writer = csv.writer(arquivo) # "W" segunifica   que é para escrita
    writer.writerow(('primeira','segunda','terceira')) #writer = escrever algum dado na documentação.
    writer.writerow((55,93,76)) #writerow = é acrescentar uma lista escrita na conjuntura do arquivo.
    writer.writerow((62,14,86))
```

```python
# Leitura de arquivos csv
with open('/content/drive/MyDrive/Colab Notebooks/PythonFundamentos-master/Cap04/Notebooks/arquivos/numeros.csv','r') as arquivo:
    leitor = csv.reader(arquivo)
    for x in leitor:
        print ('Número de colunas:', len(x))
        print(x)
```

```python
# Código alternativo para eventuais problemas com linhas em branco no arquivo
with open('/content/drive/MyDrive/Colab Notebooks/PythonFundamentos-master/Cap04/Notebooks/arquivos/numeros.csv','r', encoding='utf8', newline = '\r\n') as arquivo:
    leitor = csv.reader(arquivo)
    for x in leitor:
        print ('Número de colunas:', len(x))
        print(x)
```

```python
# Gerando uma lista com dados do arquivo csv
with open('/content/drive/MyDrive/Colab Notebooks/PythonFundamentos-master/Cap04/Notebooks/arquivos/numeros.csv','r') as arquivo:
    leitor = csv.reader(arquivo)
    dados = list(leitor)
    
    
print (dados)
```

```python
# Impriminfo a partir da segunda linha
for linha in dados[1:]:
    print (linha)
```

### Manipulando Arquivos JSON (Java Script Object Notation )
JSON (JavaScript Object Notation) é uma maneira de armazenar informações de forma organizada e de fácil acesso. Em poucas palavras, ele nos dá uma coleção legível de dados que podem ser acessados de forma muito lógica. Pode ser uma fonte de Big Data.

```python
# Criando um dicionário
dict = {'nome': 'Guido van Rossum',
        'linguagem': 'Python',
        'similar': ['c','Modula-3','lisp'],
        'users': 1000000}
```

```python
for k,v in dict.items():
    print (k,v)
```

```python
# Importando o módulo Json
import json
```

```python
# Convertendo o dicionário para um objeto json
json.dumps(dict)
```

```python
# Criando um arquivo Json
with open('/content/drive/MyDrive/Colab Notebooks/PythonFundamentos-master/Cap04/Notebooks/arquivos/dados.json','w') as arquivo:
    arquivo.write(json.dumps(dict))
```

```python
# Leitura de arquivos Json
with open('/content/drive/MyDrive/Colab Notebooks/PythonFundamentos-master/Cap04/Notebooks/arquivos/dados.json','r') as arquivo:
    texto = arquivo.read()
    data = json.loads(texto)
```

```python
print (data)
```

```python
print (data['nome'])
```

```python
# Imprimindo um arquivo Json copiado da internet
from urllib.request import urlopen
# para abrir url, vai ler, codfica para utf-8, gerando 'response'. data = conteudos de "response, lista[0]"
response = urlopen("http://vimeo.com/api/v2/video/57733101.json").read().decode('utf8') 
data = json.loads(response)[0]
```

```python
print ('Título: ', data['title'])
print ('URL: ', data['url'])
print ('Duração: ', data['duration'])
print ('Número de Visualizações: ', data['stats_number_of_plays'])
```

```python
# Copiando o conteúdo de um arquivo para outro
import os

arquivo_fonte = '/content/drive/MyDrive/Colab Notebooks/PythonFundamentos-master/Cap04/Notebooks/arquivos/dados.json'
arquivo_destino = '/content/drive/MyDrive/Colab Notebooks/PythonFundamentos-master/Cap04/Notebooks/arquivos/json_data.txt'
```

```python
# Método 1 você abrirá o arqfont com "infile", l^-ra inf como text. tendo "text", abrirá arqdes para escrita "outfile", 
#"out file" escreverá o "text" em si. ou seja arqfont ==> (copy) ==> arqdes.
with open(arquivo_fonte,'r') as infile:
    text = infile.read()
    with open(arquivo_destino,'w') as outfile:
        outfile.write(text)  
```

```python
# Método 2 Igual a cima mas, mais simples.
open(arquivo_destino,'w').write(open(arquivo_fonte,'r').read()) 
```
